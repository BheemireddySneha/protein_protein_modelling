```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt

# Load the trajectory and topology files
u = mda.Universe('/home/sneha/work/Final_models/26_12_2022/all/mtr_ab/md_protein.pdb',
                 '/home/sneha/work/Final_models/26_12_2022/all/mtr_ab/md_protein.dcd')

# Select the protein segments A and B
selection_A = u.select_atoms('segid A' and 'resnum 305')
selection_B = u.select_atoms('segid B' and 'resnum 56')

selection_C = u.select_atoms('segid C' and 'resnum 305')
selection_D = u.select_atoms('segid D' and 'resnum 56')

# Calculate the distance between two residues for every frame
distances = []
for ts in u.trajectory:
    #print(selection_A, selection_B )
    distance = np.linalg.norm(selection_A.center_of_mass() - selection_B.center_of_mass())
    distances.append(distance)

# Calculate the average distance
average_distance = np.mean(distances)

# Plot the distance as a function of time
time = u.trajectory.time
plt.plot(distances)
plt.axhline(y=average_distance, color='r', linestyle='--', label='Average')
plt.xlabel('Time (ps)')
plt.ylabel('Distance ()')
plt.title('Distance between Residue A and Residue B')
plt.legend()
plt.show()

# Calculate the distance between two residues for every frame
distances = []
for ts in u.trajectory:
    #print(selection_A, selection_B )
    distance = np.linalg.norm(selection_C.center_of_mass() - selection_D.center_of_mass())
    distances.append(distance)

# Calculate the average distance
average_distance = np.mean(distances)

# Plot the distance as a function of time
time = (u.trajectory.time)
plt.plot(distances, color='green')
plt.axhline(y=average_distance, color='r', linestyle='--', label='Average')
plt.xlabel('Time (ns)')
plt.ylabel('Distance ()')
plt.title('Distance between Residue C and Residue D')
plt.legend()
plt.show()
```
